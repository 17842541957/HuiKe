```python
import tensorflow as tf
import numpy as np
import pandas as pd
import os
from PIL import Image
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
dataset=tf.data.TFRecordDataset("/home/ma-user/work/cai.tfrecords")
def _parse(data):
    feature = {
                'label': tf.io.FixedLenFeature([], tf.int64),
                'img_raw' : tf.io.FixedLenFeature([], tf.string),
        }
    res=tf.io.parse_single_example(data,feature)
    return res
dataset2=dataset.map(_parse,num_parallel_calls=tf.data.experimental.AUTOTUNE)

img_list=[]
label=[]

for data in dataset2:
    img = tf.io.decode_raw(data['img_raw'], tf.uint8)
    #print(img.shape)
    img = tf.reshape(img,[256,256,3])
    img = tf.cast(img,dtype="float64")
    #print(data["label"])
    if data["label"]==2 or data["label"]==3:
        label.append(data["label"])
        img_list.append(img.numpy())
    else:
        pass
label=np.array(label)
img_list=np.array(img_list)
```

```python
model = tf.keras.Sequential()
#卷积1
model.add(tf.keras.layers.Conv2D(8,(4,4),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",input_shape=[256,256,3]))
#卷积2
model.add(tf.keras.layers.Conv2D(16,(4,4),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",name="conv2"))
#卷积3
model.add(tf.keras.layers.Conv2D(32,(7,7),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",name="conv3"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=1,strides=1,padding='SAME'))
model.add(tf.keras.layers.BatchNormalization())
#卷积4
model.add(tf.keras.layers.Conv2D(16,(4,4),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",name="conv4"))
#卷积4
model.add(tf.keras.layers.Conv2D(16,(4,4),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",name="conv5"))
"""
#卷积4
model.add(tf.keras.layers.Conv2D(16,(5,5),kernel_initializer="he_normal",activation="relu",strides=2,padding="same",name="conv6"))
model.add(tf.keras.layers.MaxPooling2D(pool_size=1,strides=1,padding='SAME'))
model.add(tf.keras.layers.BatchNormalization())"""


#droput层
model.add(tf.keras.layers.Flatten(name="flatten"))
model.add(tf.keras.layers.Dropout(rate=0.2,name="d3"))
#输出层
model.add(tf.keras.layers.Dense(4,activation="relu"))
model.add(tf.keras.layers.Dense(4,activation="relu"))
model.add(tf.keras.layers.Dense(4,activation="softmax"))
#模型编译
model.compile(loss="sparse_categorical_crossentropy",optimizer=tf.keras.optimizers.Adam(0.001),metrics=["accuracy"])
```
